```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import statsmodels.formula.api as smf
import math
from scipy.stats import pearsonr
```

```python
corr, p_value = pearsonr(amazon["actual_price"], amazon["rating"])
print("Correlation:", corr)
print("P-value:", p_value)
```

```python
#The Pearson correlation coefficient (r = 0.1173) indicates a very weak positive relationship between actual price and rating.
#The p-value<0.05, therefore, the result is statistically significant. 
#This suggests that expensive products do not necessarily have much higher ratings.
```

```python
# 1. Defining the split point for a 70:30 ratio
train_end = int(0.7 * len(amazon))
train_end
```

```python
#2. Splitting the dataset into Training (trn) and Testing (tst)
trn = amazon.iloc[0:train_end, ]
tst = amazon.iloc[train_end:len(amazon), ]
```

```python
# 3. Fitting the OLS Regression Model
# Defining model_1 using smf.ols with the formula string
model_1 = smf.ols(formula = "discounted_price ~ rating + rating_count", data = trn).fit()
model_1
```

```python
# 4. Printing the comprehensive statistical summary
print(model_1.summary())
```

```python
# 5. Preparing the test predictors and target variable
# Using the .loc method to separate predictors and target variable from tst
x_tst = tst.loc[:, tst.columns != "discounted_price"] 
y_tst = tst.loc[:, tst.columns == "discounted_price"]
```

```python
x_tst
```

```python
y_tst
```

```python
# 6. Generating predictions using the test set
pred_1 = model_1.predict(x_tst)
pred_1
```

```python
# 7. Calculating the residuals (Errors)
resd_1 = y_tst["discounted_price"] - pred_1
resd_1
```

```python
# 8. Visualising the Target vs Predicted values
# Using np.arange to create observation numbers for the x-axis
obsno = np.arange(0, len(y_tst), 1)
fig, ax = plt.subplots()
ax.plot(obsno, y_tst["discounted_price"], "o", label = "Target")
ax.plot(obsno, pred_1, "x", color = "red", label = "Predicted")
ax.legend(loc = "best")
plt.title("Actual vs Predicted Prices")
plt.show()
```

```python
# 9. Calculating the Root Mean Squared Error (RMSE) manually
# Following the exact step-by-step logic: power, sum, mean, then square root
se_1 = resd_1.pow(2)            # square the errors
sse_1 = se_1.sum()              # sum the squared terms
msse_1 = sse_1 / len(se_1)      # mean of squared errors
rmse_1 = math.sqrt(msse_1)
```

```python
print("The RMSE of the model is:", round(rmse_1, 2))
```
